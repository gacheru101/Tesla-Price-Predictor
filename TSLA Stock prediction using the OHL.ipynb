```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv
df = pd.read_csv("C:\\Users\\Gacheru\\Downloads\\TSLA.csv")
```

```python
#looking at the data
df.head()
#lets vizualize the close using the dates to see how TSLA has been trading in the past
df['Date'] = pd.to_datetime(df['Date'], errors = 'coerce')
sns.lineplot(x="Date",y="Adj Close",data=df)
```

```python
#select the columns that I am going to use
mydata = df[['Open','High','Low','Adj Close']]
mydata.head()

```

```python
#looking at the vizual relationship between each feature("Open,High,Low") and the label(Adj Close)
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['Open'],df['Adj Close'],'o')
axes[0].set_ylabel("Adj Close")
axes[0].set_title("Open")

axes[1].plot(df['High'],df['Adj Close'],'o')
axes[1].set_title("High")
axes[1].set_ylabel("Adj Close")

axes[2].plot(df['Low'],df['Adj Close'],'o')
axes[2].set_title("Low")
axes[2].set_ylabel("Adj Close")
plt.tight_layout();
```

```python
#separating the Feature from the label
X = df[["Open","High","Low"]]
y = df["Adj Close"]
#splitting the data by using train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#importing the linear regression and creating an object
from sklearn.linear_model import LinearRegression
```

```python
mymodel = LinearRegression()
```

```python
mymodel.fit(X_train,y_train)
```

```python
mymodel.predict(X_test)
```

```python
#cheking the score
mymodel.score(X_test,y_test)
```
